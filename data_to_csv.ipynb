```python
import json
import pandas as pd
```

```python
fp = open('top_google_charts_TH.json')
```

```python
data = json.load(fp)
```

```python
data.keys()
```

```python
len(data['app_list'])
```

```python
data = pd.DataFrame.from_dict(data['app_list'])
```

```python
list(data)
```

```python
data.drop("app_availability",axis=1,inplace=True)
data.drop("badges",axis=1,inplace=True)
data.drop("cat_int",axis=1,inplace=True)
data.drop("cat_key",axis=1,inplace=True)
data.drop("cat_keys",axis=1,inplace=True)
data.drop("cat_type",axis=1,inplace=True)
data.drop("category",axis=1,inplace=True)
data.drop("content_rating",axis=1,inplace=True)
data.drop("created",axis=1,inplace=True)
data.drop("developer",axis=1,inplace=True)
data.drop("downloads",axis=1,inplace=True)
data.drop("downloads_max",axis=1,inplace=True)
data.drop("downloads_min",axis=1,inplace=True)
data.drop("email",axis=1,inplace=True)
data.drop("from_developer",axis=1,inplace=True)
data.drop("i18n_lang",axis=1,inplace=True)
data.drop("iap_max",axis=1,inplace=True)
data.drop("iap_min",axis=1,inplace=True)
data.drop("icon_72",axis=1,inplace=True)
data.drop("market_source",axis=1,inplace=True)
data.drop("market_status",axis=1,inplace=True)
data.drop("number_ratings",axis=1,inplace=True)
data.drop("physical_address",axis=1,inplace=True)
data.drop("price",axis=1,inplace=True)
data.drop("price_i18n_countries",axis=1,inplace=True)
data.drop("price_numeric",axis=1,inplace=True)
data.drop("promo_video",axis=1,inplace=True)
data.drop("ratings_1",axis=1,inplace=True)
data.drop("ratings_2",axis=1,inplace=True)
data.drop("ratings_3",axis=1,inplace=True)
data.drop("ratings_4",axis=1,inplace=True)
data.drop("ratings_5",axis=1,inplace=True)
data.drop("version",axis=1,inplace=True)
data.drop("website",axis=1,inplace=True)
data.drop("what_is_new",axis=1,inplace=True)
```
